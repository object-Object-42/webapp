```python
import os
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()

api_key = os.getenv("OPENAI_API_KEY")
client = OpenAI()
client.api_key = api_key

# Initialize list to store dictionaries containing title and embedding
documents = []

for i in range(118):
    try:
        response = client.embeddings.create(
            input=text[i].get("Inhalt"), model="text-embedding-3-small"
        )

        doc = {"title": text[i].get("Titel"), "embedding": response.data[0].embedding}
        documents.append(doc)
    except:
        print("Error at index", i)
        print(text[i].get("Titel"))

# print(response.data[0].embedding)
print(documents[0]["title"])
print(documents[0]["embedding"])
```

```python
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

titles = [doc["title"] for doc in documents]
embeddings = [doc["embedding"] for doc in documents]
print(len(embeddings))

# Convert list of embeddings to numpy array
embeddings_array = np.array(embeddings)

# Apply t-SNE
tsne = TSNE(
    n_components=2, random_state=42, perplexity=2
)  # adjust perplexity based on your data size
embeddings_2d = tsne.fit_transform(embeddings_array)

# Create visualization
plt.figure(figsize=(10, 8))
plt.scatter(embeddings_2d[:, 0], embeddings_2d[:, 1])

# Add labels (titles) to the points
for i, title in enumerate(
    titles
):  # assuming 'titles' is a list of your document titles
    print(i, title, embeddings_2d[i, 0], embeddings_2d[i, 1])
    plt.annotate(title, (embeddings_2d[i, 0], embeddings_2d[i, 1]))

plt.title("t-SNE visualization of document embeddings")
plt.xlabel("t-SNE dimension 1")
plt.ylabel("t-SNE dimension 2")
plt.tight_layout()
plt.show()
```

```python
import pickle

# Save to disk
with open("embeddings.pkl", "wb") as f:
    pickle.dump(documents, f)
```

```python
from sqlalchemy import create_engine, MetaData, Table, select


def fetch_content(
    db_url="postgresql+psycopg://postgres:changethis@localhost:5432/app", fetch_size=5
):
    """
    Fetch content from the database.

    Args:
        db_url (str): Database connection URL
        fetch_size (int): Number of records to fetch

    Returns:
        list: List of dictionaries containing Titel and Inhalt
    """
    try:
        # Create an engine instance
        engine = create_engine(db_url, echo=True)

        # Connect to the database
        with engine.connect() as connection:
            # Define metadata and reflect the existing database schema
            metadata = MetaData()
            metadata.reflect(bind=engine)

            # Define the content table
            content_table = Table("content", metadata, autoload_with=engine)

            # Create a select statement to query the table
            stmt = select(content_table)

            # Execute the query and fetch the results
            results = connection.execute(stmt).fetchmany(size=fetch_size)

            # Process the results
            context = []
            for row in results:
                context.append(
                    {
                        "Titel": row[0],
                        "Inhalt": row[1],
                    }
                )

            return context

    except Exception as e:
        print(f"An error occurred: {str(e)}")
        return []

    finally:
        # Ensure the connection is closed
        if "connection" in locals():
            connection.close()
```
